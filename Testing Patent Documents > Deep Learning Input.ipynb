Thinking about groups of documents or patent corpuses.

We want to load the file data into the patentdocument object only when that patent document is used. The corpus generators have an issue that they can only be looped through once and because of the random aspect return a different set of documents every time.

We don't want our corpus object to store the documents (e.g. in a list). We want to store a list of references.

Think about counting case first. We want to iterate through the references, load the document as a patentdoc, compute the statistics, add the results to our running corpus totals, then delete the patentdoc and load the next one.

Can we pass a records object? Lists of publication numbers are probably our best interface. Corpus takes the numbers and converts into a list of file references. PatentCorpus object calls general interface on publications / grants passing reference details. How to combine publications and grants? Only do one set first?

Our corpus will store our dictionary of tokens (e.g. words). We will need one pass over the data to generate the dictionary.

## Reference Numerals

It would be nice to remove these to clean up our text.  

We need a function that builds a dictionary of terms and reference numbers - we can then use to remove the reference numbers and add them back in. This dictionary can also help us spot errors in our reference numbers.  

This is more complicated than it looks because what we actually have is (n-gram, reference_no) pairs.  

We also have phrases like this: "The first and second lower pivot shafts 30, 45", where we have multiple reference numerals.  

We also need to distinguish from "FIG. 9" or "there are 10 items".  

The general pattern is NOUN_PHRASE DIGIT+(,\s)?

````python
from nltk import RegexpParser, tree

def label_nounphrases(pos):
        """ Label noun phrases in the output from pos chunking. """
        grammar = '''
            NP: {<DT|PRP\$> <VBG> <NN.*>+}
                {<DT|PRP\$> <NN.*> <POS> <JJ>* <NN.*>+}
                {<DT|PRP\$>? <JJ>* <NN.*>+ }
            '''

        cp = RegexpParser(grammar)
        result = cp.parse(pos)
        ptree = tree.ParentedTree.convert(result)
        subtrees = ptree.subtrees(filter=lambda x: x.label() == 'NP')

        # build up mapping dict - if not in dict add new entry id+1;
        # if in dict label using key
        mapping_dict = {}
        pos_to_np = {}
        for st in subtrees:
            np_string = " ".join(
                [
                    leaf[0] for leaf in st.leaves()
                    if leaf[1] != ("DT" or "PRP$")
                ]
            )
            np_id = mapping_dict.get(np_string, None)
            if not np_id:
                # put ends_with here
                nps = [i[0] for i in mapping_dict.items()]
                ends_with_list = [
                    np for np in nps if ends_with(np_string, np)
                ]
                if ends_with_list:
                    np_id = mapping_dict[ends_with_list[0]]
                else:
                    np_id = len(mapping_dict)+1
                    mapping_dict[np_string] = np_id
            pos_to_np[st.parent_index()] = np_id

        # Label Tree with entities
        flat_list = []
        for i in range(0, len(ptree)):
            # print(i)
            # Label
            if isinstance(ptree[i], tree.Tree):
                for leaf in ptree[i].leaves():
                    # Unpack leaf and add label as triple
                    flat_list.append((leaf[0], leaf[1], pos_to_np.get(i, "")))
            else:
                flat_list.append(
                    (ptree[i][0], ptree[i][1], pos_to_np.get(i, ""))
                )
        return (flat_list, mapping_dict)
    
import re


def ends_with(s1, s2):
    """See if s1 ends with s2."""
    pattern = re.compile('(' + re.escape(s2) + ')$')
    located = pattern.search(s1)
    if located:
        return True
    else:
        return False
````
